```python
import matplotlib.pyplot as plt

# Dados
categories = ['-5%', '-4%', '-3%', '-2%', '-1%', 'Bal.', '1%', '2%', '3%', '4%', '5%']
values = [87, 185, 197, 75, 204, 203, 79, 190, 80, 89, 185]

# Configuração do gráfico
plt.figure(figsize=(8, 6))  # Aumenta a largura e diminui a altura
plt.bar(categories, values, color='royalblue', edgecolor='black')

# Título e rótulos
plt.title('Number of Samples After Cutoff', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('')
plt.ylabel('')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

# Exibir valores em cada barra
for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center', fontsize=15)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
